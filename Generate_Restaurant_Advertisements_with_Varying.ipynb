## Set up Gemini API

To use the Gemini API, you'll need an API key. If you don't already have one, create a key in Google AI Studio.
In Colab, add the key to the secrets manager under the "🔑" in the left panel. Give it the name `GOOGLE_API_KEY`. Then pass the key to the SDK:

```python
# Import the Python SDK
import google.generativeai as genai
# Used to securely store your API key
from google.colab import userdata

GOOGLE_API_KEY=userdata.get('GOOGLE_API_KEY')
genai.configure(api_key=GOOGLE_API_KEY)
```

Before you can make any API calls, you need to initialize the Generative Model. We will use `gemini-pro` for text generation.

```python
# Initialize the Gemini API
# Using 'gemini-pro' for text generation
gemini_model = genai.GenerativeModel('gemini-3.5-flash')
```

## Generate Restaurant Advertisements with Varying

*   List item
*   List item

Temperatures

```python
prompt = "Write an advertisement for a newly opened South Indian restaurant."

print(f"Prompt: {prompt}\n")

# Temperature = 0 (deterministic)
print("--- Generating with Temperature = 0 ---")
response_temp0 = gemini_model.generate_content(
    prompt,
    generation_config=genai.types.GenerationConfig(temperature=0)
)
print(response_temp0.text)
print("\n---\n")

# Temperature = 0.5 (balanced creativity)
print("--- Generating with Temperature = 0.5 ---")
response_temp0_5 = gemini_model.generate_content(
    prompt,
    generation_config=genai.types.GenerationConfig(temperature=0.5)
)
print(response_temp0_5.text)
print("\n---\n")

# Temperature = 1 (most creative)
print("--- Generating with Temperature = 1 ---")
response_temp1 = gemini_model.generate_content(
    prompt,
    generation_config=genai.types.GenerationConfig(temperature=1)
)
print(response_temp1.text)
print("\n---\n")
```
