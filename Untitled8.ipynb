## 1. Convert Audio into Text using Groq's Whisper Model

We'll define a function to transcribe the audio file. Make sure your `team_meeting.wav` file is accessible in the Colab environment (e.g., uploaded to your Colab session files).

```python
def transcribe_audio(audio_file_path):
    """Transcribes an audio file using Groq's Whisper model."""
    try:
        with open(audio_file_path, "rb") as file:
            transcript = client.audio.transcriptions.create(
                file=(audio_file_path, file.read()),
                model="whisper-large-v3"
            )
        return transcript.text
    except Exception as e:
        print(f"Error transcribing audio: {e}")
        return None

# Example usage: (Assuming team_meeting.wav is in the current directory)
# audio_path = 'team_meeting.wav'
# transcribed_text = transcribe_audio(audio_path)
# if transcribed_text:
#     print(f"Transcription: {transcribed_text}")
```

## 2. Extract Meeting Information

Now, we'll use a Groq language model (e.g., `llama3-8b-8192` or `mixtral-8x7b-32768`) to extract the required meeting details from the transcribed text and format them as JSON.

```python
import json

def extract_meeting_info(text):
    """Extracts meeting details from text using a Groq LLM and returns them in JSON format."""
    prompt = f"""From the following text, extract the meeting details. Ensure the output is a JSON object with the following keys: 'meeting', 'date', 'time', 'person', 'location'.

If a detail is not found, use a null value for that key.

Text: "{text}"

Example Output:
{{
  "meeting": "Project Review",
  "date": "Friday",
  "time": "3:00 PM",
  "person": "Mr. Rajesh",
  "location": "Conference Room B"
}}

JSON Output:"""

    try:
        chat_completion = client.chat.completions.create(
            messages=[
                {
                    "role": "user",
                    "content": prompt,
                }
            ],
            model="llama3-70b-8192", # Updated model to a currently supported one
            response_format={"type": "json_object"},
            temperature=0.1 # Keep it low for structured extraction
        )
        return json.loads(chat_completion.choices[0].message.content)
    except Exception as e:
        print(f"Error extracting information: {e}")
        return None
```

## Complete Workflow Example

Let's put it all together. **Remember to upload your `team_meeting.wav` file to your Colab session files before running this cell.**

```python
# Define the path to your audio file
audio_file_path = '/content/Microsoft-Teams-meeting-20250605_170332-Meeting-Recording.mp3'

# Step 1: Transcribe the audio
print("Transcribing audio...")
transcribed_text = transcribe_audio(audio_file_path)

if transcribed_text:
    print(f"Transcription: {transcribed_text}")

    # Step 2: Extract meeting information
    print("\nExtracting meeting information...")
    meeting_details = extract_meeting_info(transcribed_text)

    if meeting_details:
        print("\nExtracted Meeting Details (JSON):")
        print(json.dumps(meeting_details, indent=2))
    else:
        print("Failed to extract meeting details.")
else:
    print("Failed to transcribe audio.")
```
